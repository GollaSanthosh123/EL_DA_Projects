## import the required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### import cleaned data using pandas

```python
data=pd.read_excel("sharktankstartup.xlsx")
df=pd.DataFrame(data)
df
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ','')
df.rename(columns={'investor(s).1':'investors'},inplace=True)
df.head()
```

## Data Visualization

## *Column Chart*                                                
**Number Of Startups basedon founder background**

```python
data=df.groupby('founderbackground')['startupname'].count().reset_index()
data.rename(columns={'startupname':'numberofstartups'},inplace=True)
plt.figure(figsize=(10,6))
plt.bar(data['founderbackground'],data['numberofstartups'],color='skyblue')
plt.title("Number Of Startups basedon founder background")
plt.xlabel("FounderBackground")
plt.ylabel("No.of.Startups")
plt.xticks(rotation=45)
plt.show()
```

## *Table Chart*                                                                             
**fund received based on funding stage**

```python
data=df.groupby('fundingstage')['investmentamount(inrcr)'].sum().reset_index()
fig, ax = plt.subplots(figsize=(10, 2))
ax.axis('tight')
ax.axis('off')
table = ax.table(cellText=data.values, colLabels=data.columns, loc='center')
plt.title("Fund Received Based On Fundingstage")
plt.show()
```

## *Pie Chart*                                                                 
**investment by industry**

```python
data=df.groupby('industry')['investmentamount(inrcr)'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.pie(x=data['investmentamount(inrcr)'],labels=data['industry'],autopct='%1.1f%%',labeldistance=1.2)
plt.title('Investment By Industry')
plt.show()
```

## Average Investment By Investors

```python
data=df.groupby('investors')['investmentamount(inrcr)'].mean().reset_index()
sns.barplot(x='investmentamount(inrcr)',y='investors',data=df,color='indigo')
plt.title("Average Investment By Investors")
plt.show()
```
